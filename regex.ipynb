```python
import re

names = [
    "Francois",
    "Mike Whitney",
    "Sagun Khatri",
    "Nick Francesco",
    "Pickles",
    "K5ANR",
]


def re_names(pattern: str):
    for name in names:
        print(re.search(pattern, name))
```

```python
re_names("Fran")
```

```python
re_names("^Fran")
```
