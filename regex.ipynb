```python
import re

names = [
    "Francois",
    "Mike Whitney",
    "Sagun Khatri",
    "Nick Francesco",
    "Pickles",
    "K5ANR",
]


def re_names(pattern: str):
    for name in names:
        if match := re.search(pattern, name):
            print(match, match.group(), match.span(), match.start(), match.end())
```

```python
re_names("Fran")
```

```python
re_names("^Fran")
```

```python
re_names("i$")
```

```python
names = [
    "Brian Daugette",
    "Veronica Supersonica",
    "Tony Gasparovic",
    "Patrick Germann",
    "m!sha",
]
```

```python
def re_names(pattern: str):
    for name in names:
        if match := re.search(pattern, name):
            print(match, match.group(0))
```

```python
first_last_name = r"^(?P<fn>\w+)\s+(?P<ln>\w+)$" # uhhh too lazy to change to make accessing groups by name work
re_names(first_last_name)
```

```python
re_names(r"[a-zA-Z!]+")
```

```python
re_names(r"^[a-zA-Z!]+$")
```

```python
re_names(".+")
```

```python
re_names("^.+$") # except a newline
```

```python
def re_findall_names(pattern: str):
    for name in names:
        if match := re.findall(pattern, name):
            print(match)
```

```python
re_findall_names("[a-z]+")
```

```python
for name in names:
    for match in re.finditer("[a-z]+", name):
        print(match)
```

```python
urls = [
    "https://www.socratica.com",
    "https://www.socratica.org",
    "https://wwwwsocratica.org",
    "file://test.this.path",
    "com.socratica.www_https://",
]
```

```python
for url in urls:
    if re.match("https?", url):
        print(url)
```

```python
for url in urls:
    if re.fullmatch(r"https?://w{3}\.\w+\.(com|org)", url):
        print(url)
```
